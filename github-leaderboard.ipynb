```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
df=pd.read_csv('namelist.csv')
```

```python
df
```

```python
namelist=[]
for name in df['github_link']:
    namelist.append(name)
```

```python
namelist
```

```python
activities=[]

for url in namelist:
    r=requests.get(url).text
    data=BeautifulSoup(r,"html.parser")
    information=data.find_all('rect')
    for j in range(255,len(information)): #start from August 1st
        activity={}
        activity['url'] = url
        activity['data-date'] = information[j]['data-date']
        activity['data-count'] = information[j]['data-count']
        activities.append(activity)

# # (username, datetime, contributions)
# with open('contribution.csv','w',newline='') as f:
#     writer=csv.writer(f)
#     header=['user','datetime','contribution']
#     writer.writerow(header)
#     for i in range(0,len(activities)):
#         writer.writerow([activities[i][0],activities[i][1],activities[i][2]])
```

```python
activities
```

```python
df2 = pd.DataFrame(activities)
```

```python
df2
```

```python
df2 = df2.groupby('url')['data-count'].agg('sum').reset_index()
```

```python
df2
```

```python
df2['data-count'] = pd.to_numeric(df2['data-count'])
```

```python
df2.rename(columns={'url':'github_link'}, inplace=True)
```

```python
df2
```

```python
df3 = pd.merge(df,df2,how='left',on='github_link')
```

```python
df3.sort_values(by=['data-count'],ascending=False)
```
